$$sin(2 \pi x) * exp(-0.1t)$$  
위 식을 애니메이션으로 표현하기 위해 matplotlib의 animation을 사용하였습니다.

```
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, exp, pi
%matplotlib inline

from matplotlib import animation, rc
```

```
# 그래프 설정을 지정
fig, ax = plt.subplots()

ax.set_xlim(( 0, 1))
ax.set_ylim((-1, 1)) # 표현하고자 하는 함수의 반복 단위

line, = ax.plot([], [], lw=2)
```

```
def init():
    line.set_data([], [])
    return (line,)
```

```
def animate(t):
    x = np.linspace(0, 10, 1000)
    y = sin(2 * pi * x) * exp(-0.1 * t)
    line.set_data(x, y)
    return (line,)
```

```
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=90, interval=50, blit=True)
```

##### animation.FuncAnimation에 전달되는 인자들의 목록을 살펴보겠습니다.

* fig : 현 그림의 객체  
* init_func : 초기화 함수로써 프레임이 리셋될 때 호출됨.
* frames : 애니메이션의 초당 프레임 개수  
* interval : 프레임 간의 시간 간격 (밀리초 단위)  
* repeat : True, False 값을 가질 수 있으며 애니메이션의 반복 유무를 결정. (default는 True)
