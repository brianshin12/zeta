## 1. Markdown

### 1.1 Markdown 이란?

* Markdown 이란 2004년에 존그루버에 의해 만들어진 텍스트 기반의 마크업 언어입니다. 파일 확장자명은 .md이며 간결한 문법과 다른 플랫폼의 마크업와 변환이 쉽다는 장점을 가지고 있습니다. 
    
### 1.2 Markdown의 장.단점
    
* 장점  

    1. 문법이 간결하고 쉽다.  
    2. text형태로 저장하여 용량이 적어 보관이 용이하다.  
    3. 다양한 형태로 변환이 용이하며, 지원하는 프로그램 및 플랫폼이 다양하다.  
    
    
* 단점  

    1. 표준이 없어 사용자마다 문법이 상이하다.  
    2. 모든 HTML마크업을 대신하지 못한다.

## 2. Markdown의 문법

### 2.1 글머리
    
* Markdown의 글머리는 글자 앞 #의 갯수로 글자의 크기를 1~6으로 나눌 수 있다.  
```
# This is H1
## This is H2
### This is H3
#### This is H4
##### This is H5
###### This is H6
```  
# This is H1
## This is H2
### This is H3
#### This is H4
##### This is H5
###### This is H6

### 2.2 강조

* *이텔릭체*를 사용하기 위해서는 이텔릭체로 표시하고 싶은 글 앞뒤로 별표```*``` 혹은 언더바```_```를 쓰면 됩니다.
```
*single asterisks*
_single underscores_
```  
*single asterisks*  
_single underscores_

* **두껍게**를 사용하기 위해서는 두껍게 표시하고 싶은 글 앞뒤로 별표 두개 ```**``` 혹은 언더바 두개 ```__```를 쓰면 됩니다.
```
**double asterisks**
__double asterisks__
```  
**double asterisks**  
__double underscores__
   
* <u>밑줄</u>을 사용하기 위해서는 밑줄을 치고 싶은 글 앞에 ```<u>``` 뒤에 ```</u>```를 쓰면 됩니다.
```
<u>underline</u>
```  
<u>underline</u>

* ~~취소선~~을 사용하기 위해서는 취소선을 치고 싶은 글 앞뒤에 물결 두개```~~```를 쓰면 됩니다.
```
~~cancelline~~
```  
~~cancelline~~

### 2.3 목록
* Markdown에서 목록을 정리하기 위해 쓰이는 방법은 크게 두 가지가 있는데 하나는 순서가 있는 목록이고 다른 하나는 순서가 없는 목록입니다.  

    순서가 있는 목록은 숫자와 점을 이용하여 표시하는데 이 때 어느 번호를 입력하더라도 내림차순으로 자동으로 정리가 됩니다.

```
1. 첫번째
3. 두번째
2. 세번째
```
    
1. 첫번째
3. 두번째
2. 세번째

    순서가 없는 목록은 * , + , - 표를 이용하여 표시합니다.

```
* 순서가 없는 목록1
    + 순서가 없는 목록2
        - 순서가 없는 목록3
```

* 순서가 없는 목록1  
    + 순서가 없는 목록2  
        - 순서가 없는 목록3

### 2.4 수평선
* ```***``` , ```---``` , ```-----------```등을 이용하여 수평선을 나타낼 수 있다.
***
---
-----------

### 2.5 링크
* 문서 내 일반 URL이나 걲쇠괄호```<>```안의 URL은 자동으로 링크를 형성합니다.
```
www.naver.com
<https://www.google.com>
```

www.naver.com  
<https://www.google.com>  
* 참조링크를 이용하여 링크를 형성할 수도 있습니다.  

```
[참조링크]:www.naver.com
```  

[참조링크]: <https://www.naver.com>
 
 [참조링크]
 
### 2.6 이미지
* 문서 내 이미지를 넣는 법은 ```![그림제목](그림의 링크)```를 걸면 됩니다.  
    예를 들면 다음과 같습니다.
``` 
![공정제어](https://www.scienceall.com/nas/image/201008/20100804_9MSBM386.jpg) 
```
![공정제어](https://www.scienceall.com/nas/image/201008/20100804_9MSBM386.jpg)  
***
## 3 마무리
* 지금까지 markdown의 기본적인 문법 및 사용법을 알아 보았습니다. markdown이 가지고 있는 여러가지 장점 중 제가 생각했을 때 가장 큰 장점은 쉬운 문법인 것 같습니다.Google에 markdown 사용법을 검색하면 수 많은 잘 정리된 글들이 많아 markdown에 조금만 관심을 가지면 누구다 쉽게 접근이 가능하고 하루 정도만 투자한다면 쉽고 편리하게 잘 정돈된 텍스트 문서를 작성할 수 있을 것 같습니다.
